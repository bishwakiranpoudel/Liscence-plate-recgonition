```python
import splitfolders
import os

source_dir = os.path.abspath('./vehicle_number_plate_detection')
destination_dir = os.path.abspath('data/')

splitfolders.ratio(source_dir, destination_dir, seed=42, ratio=(0.8, 0.2))

print("Dataset split completed")
```

```python
from ultralytics import YOLO
import torch

model = YOLO('./best.pt')
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
model.to(device)
```

```python
results = model.train(
    model=None,
    data='./number_plate_data.yaml',
    imgsz=640,
    epochs=40,
    batch=64,
    project='training_results',
    name='license_plate_detection_yolov8_mar21',

)
```
